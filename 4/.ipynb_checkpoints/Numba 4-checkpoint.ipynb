```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math


def clip(i):
    return np.clip(i, 0, 1)

def show(A): 
    plt.imshow(cv2.cvtColor(A, cv2.COLOR_BGR2RGB) * 255)
    
```

```python
print('Reading images')

post = cv2.imread('poster.jpeg')
fram = cv2.imread('frame.jpeg')
plt.imshow(fram)
```

```python
orb = cv2.ORB_create()

k1, d1 = orb.detectAndCompute(post, None)
k2, d2 = orb.detectAndCompute(fram, None)

k1 = np.array(k1)
k2 = np.array(k1)

plt.imshow(d1)
cv2.imwrite('test.jpeg', d1)

print(len(k1), len(k2))
print(post.shape[0], fram.shape[0])
a = post.shape[0]
b = fram.shape[0]
height = max(a, b)
width = post.shape[1]+ fram.shape[1]
img = np.zeros((height, width, 3), np.uint8)

dx = post.shape[1]


# comb = np.concatenate((post, fram), axis=1)

img[:post.shape[1], :dx] = k1.copy()
img[:fram.shape[1], dx:] = k2.copy()
```

```python
post2 = cv2.imread('poster.jpeg')
fram2 = cv2.imread('frame.jpeg')
gray=cv2.cvtColor(post2, cv2.COLOR_BGR2GRAY)

out1 = cv2.drawKeypoints(fram, k1, None, color=(0,255,0))
out2 = cv2.drawKeypoints(post, k2, None, color=(0,255,0))

cv2.imwrite('test1.jpeg', out1)
cv2.imwrite('test2.jpeg', out2)
plt.imshow(out2)
```
